# Let's start preparing the titanic dataset for training!

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
dataFrame = pd.read_csv('https://raw.githubusercontent.com/gitCodingMan/Titanic-dataset-preperation/master/train.csv')

print("Features: ", dataFrame.columns)
print("Size: ", dataFrame.size)
```

Take 10 concrete examples of the the data to see how it actually looks

```python
dataFrame[:10]
```

```python
# Reorganize the data. I will left out the passengers names, the cabin, ticket and the passengers id. To me it looks like they have no influence.
# Moreover I will convert the sex in a boolean value(0 = Female, 1 = Male).

try:
    dataFrame = dataFrame.drop(columns=['PassengerId', 'Name', 'Cabin', 'Ticket'])
except KeyError:
    # Columns are already deleted
    pass

dataFrame[:10]
```

# Delete all rows containing emtpy values and change some datatypes

```python
dataFrame.dropna(inplace=True)

# Round the 'Fare' column
dataFrame['Fare'] = dataFrame['Fare'].round(0)
dataFrame['Fare'] = dataFrame['Fare'].astype(int)

# Convert object over boolean to int
dataFrame['Sex'] = dataFrame['Sex'].astype(bool)
dataFrame['Sex'] = dataFrame['Sex'].astype(int)

# Convert 'Age' to an int
dataFrame['Age'] = dataFrame['Age'].astype(int)

for column in dataFrame:
    print(column, '-', dataFrame[column].dtype)

dataFrame[:10]
```

Also I will replace the 'Embarked' values by integers, because it seems like they are always S, C or Q.

```python
print(dataFrame['Embarked'].unique())

dataFrame['Embarked'].replace({'S': 0, 'C': 1, 'Q': 2}, inplace=True)

print('Embarked', '-', dataFrame['Embarked'].dtype)
```
